```python
import requests

# Fetch a web page
r = requests.get("https://medium.com/@ebaad4000/diabetes-detection-analysis-through-ml-in-data-science-f7753b586156")
print(r.text)
```

```python
import re

# Remove HTML tags using RegularExpression
pattern = re.compile(r'<.*?>')  # tags look like <...>
print(pattern.sub('', r.text))  # replace them with blank
```

```python
from bs4 import BeautifulSoup
```

```python

# Remove HTML tags using Beautiful Soup library
soup = BeautifulSoup(r.text, "html5lib")
print(soup.get_text())
```

```python
#mw-content-container
# Find all articles
summaries = soup.find_all("div", class_="a b c")
summaries[0]
```
